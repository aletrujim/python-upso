**Diagramas de dispersión o puntos**

Un gráfico de dispersión muestra en un eje cartesiano la relación que existe entre dos variables.

scatter(x, y): Dibuja un diagrama de puntos con las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y.

```python
fig, ax = plt.subplots()

# Dibujar puntos
ax.scatter(x = [1, 2, 3], y = [3, 2, 1])

# Guardar el gráfico en formato png
plt.savefig('primerafigura.png')

# Mostrar el gráfico
plt.show()
```

**Diagramas de líneas:**

Un gráfico de líneas es una representación gráfica en un eje cartesiano de la relación que existe entre dos variables reflejando con claridad los cambios producidos.

plot(x, y): Dibuja un polígono con los vértices dados por las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y.

```python
import numpy as np

# Dibujar una linea 
x = np.linspace(0, 10, 20)  
print(x)

y = 2 * x
print(y)

plt.plot(x, y, 'r') #b-, r--, *

plt.show()
```

**Cambiar el aspecto de los gráficos:**

Colores, marcadores, estilo de linea, titulos, ejes, leyenda, grilla

```python
# genera 20 números aleatorios
temperaturas = {'Bahia Blanca': 1 - np.random.randn(20), 
                'Punta Alta': 2 + np.random.randn(20),
                'Puerto Madryn': 4 + np.random.randn(20)}

# Creamos la gráfica
fig, ax =  plt.subplots(figsize=(10, 3))

print(temperaturas['Bahia Blanca'])

# Pintamos cada una de las series
ax.plot(temperaturas['Bahia Blanca'], color = 'r', marker = 'o', label = 'Bahia Blanca')
ax.plot(temperaturas['Punta Alta'], color = 'green', marker = '*', label = 'Punta Alta')
ax.plot(temperaturas['Puerto Madryn'], color = 'blue', label = 'Puerto Madryn')

# Añadimos Formato al gráfico
# Titulo
ax.set_title('Ejemplo de gráfico de Lineas')
# Eje X
ax.set_xlabel('Numero de elemento de la serie')
# Eje Y
ax.set_ylabel('Valor del elemento')

# Grilla / cuadricula
ax.grid(True)

# Leyenda
ax.legend(loc='upper right')

plt.show()
```

**Diagramas de areas:**

fill_between(x, y): Dibuja el area bajo el polígono con los vértices dados por las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y. 

```python
fig, ax = plt.subplots()
ax.fill_between([1, 2, 3, 4], 
                [1, 2, 0, 0.5], 
                color='red')
plt.show()
```

**Diagramas de barras: Verticales y horizontales**

Un gráfico de barras es una representación gráfica en un eje cartesiano de las frecuencias de una variable cualitativa o discreta.

bar(x, y): Dibuja un diagrama de barras verticales donde x es una lista con la posición de las barras en el eje X, e y es una lista con la altura de las barras en el eje Y. 

```python
fig, ax = plt.subplots()
ax.bar([1, 2, 3], [3, 2, 1])

# Titulo del grafico 
plt.title('Mi primer diagrama de barras')

plt.show()
```

**Múltiples gráficos:**

Es posible dibujar varios gráficos en distintos ejes en una misma figura organizados en forma de tabla. Para ello, cuando se inicializa la figura y los ejes, hay que pasarle a la función subplots el número de filas y columnas de la tabla que contendrá los gráficos.

Si se quiere que los distintos ejes compartan los mismos límites para los ejes se pueden pasar los parámetros sharex = True para el eje x o sharey = True para el eje y.

```python
fig, ax = plt.subplots(2, 3, sharey = True, figsize=(10, 10))

dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Bahia Blanca':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                'Salta':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25],
                'Esquel':[25.7, 29.3, 2.6, 31, 30.2, 26.7, 29],
                'Tucuman':[31.2, 27.6, 29, 28.9, 31, 27.8, 28],
                'Nueva':[31.2, 27.6, 29, 28.9, 31, 27.8, 28]}

ax[0, 0].bar(dias, temperaturas['Bahia Blanca'])
ax[0, 1].plot(dias, temperaturas['Salta'], '--', color = 'tab:green')
ax[1, 0].plot(dias, temperaturas['Esquel'])
ax[1, 1].fill_between(dias, temperaturas['Tucuman'], color = 'tab:orange')
ax[1, 2].plot(dias, temperaturas['Nueva'])

plt.show()
plt.savefig('nombre.png')
```
